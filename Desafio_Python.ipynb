```
#Controle folha de pagamento
#Fornecer menu com 1) Cadastrar funcionário e 2) Imprimir contracheque
#1) Deve ser solicitado que digite o nome e o salário bruto do funcionário
#   --> Funcionário deve ser armazeado em lista
#   --> Pode armazenar em memória ou arquivo
#2) Deve ser solicitado que digite o índice do funcionário que ele deseja imprimir no contracheque
#   --> Listas começam em zero
#   --> Após o usuário digitar o índice, programa deve realizar calcular o desconto do INSS e do IRRF, sem considerar variáveis adicionais
#Primeiro cálculo do INSS

lista_funcionario = []
lista_salario = []

valor = 0
inss = 0
irrf = 0

##INSS

aliquota1 = 0.075
aliquota2 = 0.09
aliquota3 = 0.12
aliquota4 = 0.14
aliquota5 = 713.10

valor_limite1 = 1045
valor_limite2 = 2089.60
valor_limite3 = 3134.40
valor_limite4 = 6101.06
valor_limite5 = 6101.07

##IRRF

irrf_aliquota0 = 0
irrf_aliquota1 = 0.075
irrf_aliquota2 = 0.15
irrf_aliquota3 = 22.5 / 100
irrf_aliquota4 = 27.5 / 100

irrf_valor_limite0 = 1903.98
irrf_valor_limite1 = 2826.65
irrf_valor_limite2 = 3751.05
irrf_valor_limite3 = 4664.68
irrf_valor_limite4 = 4664.69

valor_dedutivel0 = 0
valor_dedutivel1 = 142.80
valor_dedutivel2 = 354.80
valor_dedutivel3 = 636.13
valor_dedutivel4 = 869.36

opcao = input("\nSelecione uma opção abaixo: \n\n1 - Cadastrar funcionário\n2 - Imprimir contracheque\n3 - Sair\n")

while (not opcao == "3"):

  if opcao == "1":
    nome = input("\nQual o nome do funcionário? ")
    lista_funcionario.append(nome)
    valor = float(input("Qual o salário bruto do funcionário? "))
    lista_salario.append(valor)

    opcao = input("\nSelecione uma opção abaixo: \n\n1 - Cadastrar funcionário\n2 - Imprimir contracheque\n3 - Sair\n")

    if opcao == "2":

      indice = int(input("Digite o índice do funcionário: \n"))
      valor = lista_salario[indice]
    if (valor <= valor_limite1):
      inss = valor * aliquota1
      print("Desconto do INSS R$ {0:.2f}".format(inss))

    elif (valor <= valor_limite2):
      inss = valor_limite1 * aliquota1  
      inss += (valor - valor_limite1) * aliquota2
      print("Desconto do INSS R$ {0:.2f}".format(inss))
      
    elif (valor <= valor_limite3):
      inss = valor_limite1 * aliquota1  
      inss += (valor_limite2 - valor_limite1) * aliquota2
      inss += (valor - valor_limite2) * aliquota3
      print("Desconto do INSS R$ {0:.2f}".format(inss))
        
    elif (valor <= valor_limite4):
      inss = valor_limite1 * aliquota1  
      inss += (valor_limite2 - valor_limite1) * aliquota2
      inss += (valor_limite3 - valor_limite2) * aliquota3
      inss += (valor - valor_limite3) * aliquota4
      print("Desconto do INSS R$ {0:.2f}".format(inss))

    elif (valor > valor_limite5):
      inss = aliquota5
      print("Desconto do INSS R$ {0:.2f}".format(inss))
      
    #else:
    #  inss = valor - aliquota5
    #  print("Desconto do INSS R$ {0:.2f}".format(aliquota5))
      
    total = valor - inss
    print("Salário com o desconto INSS R$ {0:.2f}".format(total))

    if (total <= irrf_valor_limite0):
      print("Isento de desconto do IRRF") 

    elif (total <= irrf_valor_limite1):
      irrf = total * irrf_aliquota1 - valor_dedutivel1
      print("Desconto IRRF R$ {0:.2f}".format(irrf))
      print("Salário Líquido R$ {0:.2f}".format(total - irrf))

    elif (total <= irrf_valor_limite2):
      irrf = total * irrf_aliquota2 - valor_dedutivel2
      print("Desconto IRRF R$ {0:.2f}".format(irrf))
      print("Salário Líquido R$ {0:.2f}".format(total - irrf))

    elif (total <= irrf_valor_limite3):
      irrf = total * irrf_aliquota3 - valor_dedutivel3
      print("Desconto IRRF R$ {0:.2f}".format(irrf))
      print("Salário Líquido R$ {0:.2f}".format(total - irrf))

    elif (total >= irrf_valor_limite4):
      irrf = total * irrf_aliquota4 - valor_dedutivel4
      print("Desconto IRRF R$ {0:.2f}".format(irrf))
      print("Salário Líquido R$ {0:.2f}".format(total - irrf))
    opcao = input("\nSelecione uma opção abaixo: \n\n1 - Cadastrar funcionário\n2 - Imprimir contracheque\n3 - Sair\n")

print("\n########Programa finalizado!############")
    
```
